```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame
```

```python
df = pd.read_csv('Test3_Full_Original.csv',
            delimiter=',', names=['x', 'y', 'Magicv2', 'localization4', 'localizationlab', 'Magicv2_5g', 'UVIC_Eduroam1', 'UVIC_Eduroam2', 'localization3', 'localization_lab5g', 'localization1', 'localization4_5g', 'UVIC_Eduroam3', 'localization2', 'localization3_5g', 'UVIC_Eduroam4', 'localization1_5g'])
df = df.loc[:,['Magicv2', 'localization4', 'localizationlab', 'Magicv2_5g', 
       'localization3', 'localization_lab5g', 'localization1', 'localization4_5g', 'localization2',
       'localization3_5g', 'localization1_5g']]
```

```python
#Get 11 Zscores for the 11 fingerprints
df['zscore1'] = (df.Magicv2 - df.Magicv2.mean())/(df.Magicv2.std())
df['zscore2'] = (df.localization4 - df.localization4.mean())/(df.localization4.std())
df['zscore3'] = (df.localizationlab - df.localizationlab.mean())/(df.localizationlab.std())
df['zscore4'] = (df.Magicv2_5g - df.Magicv2_5g.mean())/(df.Magicv2_5g.std())
df['zscore5'] = (df.localization3 - df.localization3.mean())/(df.localization3.std())
df['zscore6'] = (df.localization_lab5g - df.localization_lab5g.mean())/(df.localization_lab5g.std())
df['zscore7'] = (df.localization1 - df.localization1.mean())/(df.localization1.std())
df['zscore8'] = (df.localization4_5g - df.localization4_5g.mean())/(df.localization4_5g.std())
df['zscore9'] = (df.localization2 - df.localization2.mean())/(df.localization2.std())
df['zscore10'] = (df.localization3_5g - df.localization3_5g.mean())/(df.localization3_5g.std())
df['zscore11'] = (df.localization1_5g - df.localization1_5g.mean())/(df.localization1_5g.std())
```
